```python
#Computing Pairwise Correlation of Columns
cor=df.corr()
plt.figure(figsize=(12,10))
sns.heatmap(cor, annot=True, cmap='coolwarm')
plt.show()
```

#### Correlation analysis
- if corr(x,y)>+0.5 = very good / strong
- if corr(x,y) blw +0.1 to +0.5 = slightly good 
- if corr(x,y)<-0.5 = very good / strong
- if corr(x,y) blw -0.1 to -0.5 = slightly good 
- if corr(x,y) is /w -0.1 to +0.1 = bad/ weak
        - bad means either non-linear , or no-relation

Analytics:
- Quantity vs Sales= 0.2 has slightly good correlation
- Profit vs Sales = 0.48 has slightly good correlation
- Discount vs Sales = -0.029 has bad correlation

### 4. Exploratory Data Analytics

#### 4.1 Data visualization

```python
df.columns
```

From the above columns, the ['Ship Mode', 'Segment', 'City', 'State', 'Region', 'Category','Sub-Category'] are the "categorical" data, while the rest are the "numeric" data

```python
#Plot the pairplot
sns.set_palette('viridis')
sns.pairplot(df)
plt.show()
```

```python
# To visualize the pair plot one-by-one (you can do it with others variables(optional))

plt.figure(figsize=(12,8))
sns.scatterplot(x='Profit', y='Quantity', hue='Ship Mode', data=df)
plt.title ('Ship Mode vs Profit vs Quantity', fontsize=25)
plt.show()
```

```python
# Categorical - barplot (from seaborn use countplot)
import matplotlib.ticker as ticker
categorical = ['Ship Mode', 'Segment', 'Region','State', 'Category', 'Sub-Category']
ncount = len(categorical)


for col in categorical:
    sns.set_context(font_scale=2, rc={'font.size':10, 'axes.labelsize':20})
    plt.figure(figsize=(17,6))
    ax = sns.countplot(df[col])
    sns.set_style('whitegrid')
    for p in ax.patches:
        ax.annotate('{:.2f}'.format(p.get_height()), (p.get_x()+0.15, p.get_height()+5))
    plt.rc('xtick', labelsize=15)    # fontsize of the tick labels
    plt.xticks(rotation=90)
    plt.rc('ytick', labelsize=15)    # fontsize of the tick labels
    plt.show()
    
```

```python
#Selling Categories
df_cats = df.groupby('Category')['Quantity'].sum().reset_index()
df_cats
```

### Analyze the profit

```python
#Most Profit product
dfprofit_product = df.groupby('Sub-Category')['Profit'].sum().reset_index().sort_values(by='Profit',ascending=False)
dfprofit_product.reset_index(drop=True,inplace=True)
dfprofit_product = dfprofit_product.head(10)
dfprofit_product
```

```python
#Most Profit cities

dfprofit_cities = df.groupby('City')['Profit'].sum().reset_index().sort_values(by='Profit',ascending=False)

dfprofit_cities = dfprofit_cities.head(10)
dfprofit_cities
```

```python
#For create the scatter plot 

sns.lineplot((x='Discount',y='Profit',data=df)
plt.legend('Profit')
#df.plot.scatter(x='Discount',y='Profit',c='red',s=50,figsize=(12,5))
#plt.show()
```

```python
#3D Visualization

df[['Discount','Profit']].iplot(kind='spread')
```

```python
#Quantities Ordered of each Segment
df_group = df.groupby('Segment')['Quantity'].sum().reset_index()
labels = df['Segment'].unique()
colors = ['lightblue', 'lightgreen', 'lightyellow', 'orange']
plt.figure(figsize=(5,5))
plt.pie(df_group['Quantity'],autopct='%1.1f%%',labels=labels,explode=(0.02,0.02,0.02), colors=colors)
plt.title('Ordered quantities of each segment',size= 15)
plt.figure(0)

#Quantities Ordered of each region
df_region=df.groupby('Region')['Quantity'].sum().reset_index()
labels = df_region['Region'].unique()
plt.figure(figsize=(5,5))
plt.pie(df_region['Quantity'],autopct='%1.1f%%',labels=labels, explode=(0.02,0.02,0.02,0.02),colors=colors)
plt.title('Ordered quantities of each region',size=15)
plt.figure(1)

#Distribution of profits categorywise
dfprofit = df.groupby('Category')['Profit'].sum().reset_index()
plt.figure(figsize=(5,5))
labels=dfprofit['Category'].unique()
plt.pie(dfprofit['Profit'],autopct='%1.1f%%',labels=labels,explode=(0.02,0.02,0.02), colors=colors)
plt.title('Distribution of profits categorywise',size=15)
plt.figure(2)

#Distribution of sold products 
plt.figure(figsize=(5,5))
labels=df_cats['Category'].unique()
plt.pie(df_cats['Quantity'],autopct='%1.1f%%',labels=labels,explode=(0.02,0.02,0.02), colors=colors)
plt.title('Distribution of sold products',size=15)
plt.figure(3)

plt.show()
```

Based on the above data, we could achieve our goal, which is "find out the weak areas where you can work to make more profit":

- The consumer segment has the highest order, 51.5 compare to the corporate and home office.
- The most profit distribution category comes from the Technology segment, where the lowest one is from the Furniture segment. However, the most distributed products have come from the office supplies 60.4%, whereas the Technology segment has the most insufficient distribution.
- Most of the segments (Consumer, Corporate, and Home office) were preferred to use standard shipping mode. This condition happened because the cost for standard shipping is much lower than that of same-day shipping, which has the least count. We can see that the delivery time does not have a significant effect.  
- Based on the above (Discount vs. profit) graph, the increase in Discounts will decrease the profit where the most profit value comes from the product with a 0% discount.
